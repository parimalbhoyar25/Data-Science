# Merging DataFrame

```python
d1=DataFrame({'key':['x','z','y','z','x','x'],'Dataset1':[0,1,2,3,4,5]},index=[0,1,2,3,4,5])
```

```python
d1
```

```python
d2=DataFrame({'key':['q','y','z'],'Dataset2':[1,2,3]},index=[0,1,2])
```

```python
d2
```

```python
pd.merge(d1,d2)
```

```python
pd.merge(d1,d2,on='key') # ON keyword is used for select columns for merging
```

```python
pd.merge(d1,d2,how='left') # here the reference column for merging will be used by left column wise
```

```python
pd.merge(d1,d2,how='right') # here the reference column for merging will be used by right column wise
```

```python
pd.merge(d1,d2,how='outer') # here the reference column for merging will be used by right and left column wise
```

```python
d3=DataFrame({'Dataset3':[0,1,2,3,4],'key':['y','y','x','x','z']})
```

```python
d3
```

```python
d4=DataFrame({'Dataset2':[0,1,2,3,4,5],'key':['x','x','x','y','q','q']})
```

```python
d4
```

```python
pd.merge(d3,d4)
```

```python
pd.merge(d3,d4,how='left')
```

```python
pd.merge(d3,d4,on='key')
```

```python
pd.merge(d3,d4,how='right')
```

```python
pd.merge(d3,d4,how='outer')
```

```python
df_left=DataFrame({'key':['x','x','y','y','z'],'data':[0,1,2,3,4]})
```

```python
df_left
```

```python
df_right=DataFrame({'group_data':[10,20]},index=['x','x'])
```

```python
df_right
```

```python
pd.merge(df_left,df_right,left_on='key',right_index=True,how='outer') 
# left df ,right df,left df key column,right df index,method
```

```python
pd.merge(df_left,df_right,left_on='key',right_index=True,how='right')
```
