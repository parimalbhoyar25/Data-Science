# Bivariate graph analysis

### lm Plot

```python
#plotting graph between total_bill and tip attribute
sns.lmplot(x='total_bill',y='tip',data=df1)      #lm=linear model
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=df1,fit_reg=False)   #fit_reg=False removes the regression line from graph
plt.show()
```

## Multivariate graph analysis 

when more than 2 variables are used

```python
sns.lmplot(x='total_bill',y='tip',hue='sex',data=df1,fit_reg=False)
plt.show()
#ity shows graph between total_bill and tip while classifying with sex column.
```

```python
sns.lmplot(x='total_bill',y='tip',hue='smoker',data=df1,fit_reg=False)
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',hue='smoker',markers=['x','o'],data=df1,fit_reg=False)  #to give diffrent marker shape
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',hue='smoker',markers=['x','o'],palette='winter',data=df1,fit_reg=False) #to give diffrent color
plt.show()
```

Now we can also arrange our graph in Rows and Columns or both by using row and column parameters as shown below

```python
sns.lmplot(x='total_bill',y='tip',row='day',data=df1,fit_reg=False)
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',col='time',data=df1,fit_reg=False)
plt.plot()
```

```python
sns.lmplot(x='total_bill',y='tip',row='smoker',col='time',data=df1,fit_reg=False)
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',row='smoker',col='time',hue='sex',data=df1,fit_reg=False)
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',col='day',hue='sex',data=df1,fit_reg=False)
plt.show()
```

### Stripplot

```python
sns.stripplot(x='day',y='total_bill',data=df1)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=df1,jitter=False)    #compress the spread
plt.show()
```

### Swarmplot

```python
sns.swarmplot(x='day',y='total_bill',data=df1)
plt.show()
```
