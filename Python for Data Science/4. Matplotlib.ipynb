To create sin and cosine wave

```python
x1=np.arange(0,12,0.1)
y1=np.sin(x1)
x2=np.arange(0,12,0.1)
y2=np.cos(x2)
```

```python
plt.figure(figsize=(10,5))
plt.plot(x1,y1,color='r',markersize=10,label='first line',linestyle='-',linewidth=3)
plt.plot(x2,y2,color='b',markersize=10,label='second line',linestyle='-',linewidth=3)
plt.title('line graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid()
plt.legend()
plt.show()
```

# Bar Graph

```python
x1=[3,5,7,9]
y1=[12,15,16,8]
x2=[4,6,8,10]
y2=[15,17,19,10]
```

```python
plt.bar(x1,y1,label='data1')
plt.bar(x2,y2,label='data2')
plt.title('Bar Graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

### Horizantal Bar Graph

```python
plt.barh(x1,y1,label='data1')
plt.barh(x2,y2,label='data2')
plt.title('Bar Graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

```python
df=DataFrame({'Revenue(billons$)':[800,200,500,600,1000,300]},index=['TCS','Wipro','Infosis','Google','Fireblaze','yahoo'])
```

# Histogram

```python
bsm=[113,85,90,150,149,88,93,115,135,80,77,82,129]
```

```python
plt.hist(bsm)
plt.show()
```

```python
plt.hist(bsm,rwidth = 0.9) #by default width is 1
plt.show()
```

```python
#plt.figure(figsize=(10,8))
plt.hist(bsm,bins=4,rwidth=0.9)
plt.show()
```

```python
plt.hist(bsm,bins=[80,100,135,150],rwidth=0.9) #for check distribution
plt.show()
```

```python
plt.hist(bsm,bins=[80,100,120,140,160],rwidth=0.9) #for check distribution
plt.show()
```

```python
bsf=[67,98,89,120,133,150,84,69,89,79,120,112,100]
```

```python
plt.hist([bsm,bsf],rwidth=0.9,label=['Male','Female'],bins=[70,100,125,150],color=['r','b'])
plt.legend()
plt.show()
```

# Pie Chart

```python
Expencevalues = [1400,600,300,410,250]
Expencelabel= ['rent','food','recharge','petrol','other']
```

```python
plt.figure(figsize=(10,5)) #for size
plt.pie(Expencevalues, labels=Expencelabel,autopct='%1.2f%%') #autopct for float
plt.axis('equal') #for proper data distribute
plt.show()
```
